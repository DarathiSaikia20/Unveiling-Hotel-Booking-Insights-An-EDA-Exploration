## 3. ***Data Wrangling***

````python
# Creating a copy of the original dataset for analysis
df_copy=df.copy()
````

````python
# Write your code to make your dataset analysis ready.
# Dropping the 'company' and 'agent' columns due to a significant number of missing values
df_copy.drop(['company','agent'],axis=1,inplace=True)
df_copy.columns #  Viewing the columns
````

````python
# Filling null values in numeric columns with median and in text columns with mode
df_copy['children'] = df_copy['children'].fillna(df_copy['children'].median())
df_copy['country'] = df_copy['country'].fillna(df_copy['country'].mode()[0])
````

````python
# Taking a first look at the copied dataset
df_copy.head()
````

````python
# After removing and filling the null values, checked for any remaining missing values
df_copy.isnull().sum()
````

````python
# Changing the data type of the "children" column from float to int64 as it is more appropriate for this data
df_copy['children'] = df_copy['children'].astype('int64')
````

````python
# Displaying dataset information to verify the change of data type
df_copy.info()
````

````python
# Describing the copied dataset
df_copy.describe()
````

````python
# Introducing New Columns to the Dataset
# Creating a new 'arrival_date' column by combining day, month, and year, then converting it to datetime
df_copy['arrival_date'] = df_copy['arrival_date_day_of_month'].astype(str) + '-' + df_copy['arrival_date_month'] + '-' + df_copy['arrival_date_year'].astype(str)
df_copy['arrival_date'] = pd.to_datetime(df_copy['arrival_date'])

# Creating a new column 'total_stays' for each booking
df_copy['total_stays'] = df_copy['stays_in_weekend_nights'] + df_copy['stays_in_week_nights']

# Creating a new column 'revenue' for analysis by multiplying average daily rate (adr) with total stays
df_copy['revenue'] = df_copy['adr'] * df_copy['total_stays']

# Creating a new column 'reserved_room_assigned' based on a comparison of reserved and assigned room types
df_copy['reserved_room_assigned'] = df_copy.apply(lambda x: 'Assigned' if x['reserved_room_type'] == x['assigned_room_type'] else 'Not Assigned', axis=1)

# Creating a new column 'Total_person' as the sum of total adults, children, and babies
df_copy['Total_person'] = df_copy['adults'] + df_copy['children'] + df_copy['babies']

# Creating a new column 'customer_category' based on the total number of persons, categorizing as single, couple, or family
df_copy['customer_category'] = df_copy['Total_person'].apply(lambda x: 'Single' if x == 1 else ('Couple' if x == 2 else 'Family'))
````

````python
# Displaying the count of rows and columns in the copied dataset
df_copy.shape
````

````python
# Extracting the month from the 'arrival_date' column and counting the bookings for each month
monthly_booking = df_copy['arrival_date_month'].value_counts().sort_index()
monthly_booking
````

````python
# Calculating the count of bookings for each hotel
booking_by_hotel=df_copy.groupby(['hotel']).size().reset_index(name='Number of bookings')
booking_by_hotel
````

````python
# Calculating booking cancellations for each hotel
cancelation_by_hotel=df_copy.groupby('hotel')['is_canceled'].mean().reset_index(name='cancellation_rate')
cancelation_by_hotel
````

````python
# Calculating total revenue for each hotel
revenue_by_hotel = df_copy.groupby('hotel')['revenue'].sum().reset_index(name='Total Revenue')
revenue_by_hotel
````

````python
# Calculating the count of bookings for each country
booking_count_by_country = df_copy['country'].value_counts().reset_index()
booking_count_by_country.columns = ['country', 'booking_count']

# Sorting the countries based on the booking count in descending order
top_10_countries = booking_count_by_country.head(10)

# Displaying the top 10 countries with the highest booking count
top_10_countries
````

#### Chart - 1 : Count Plot for Bivariate Analysis of Hotel Bookings

````python
# Chart - 1 visualization code
# Visualize the monthly booking trends, including cancellations, to comprehend the fluctuations in the number of bookings across different months.

# Define the order of months
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
# Convert 'arrival_date_month' to categorical with specified order

df_copy['arrival_date_month'] = pd.Categorical(df_copy['arrival_date_month'], categories=month_order, ordered=True)

# Set the figure size and style

plt.figure(figsize=(8, 4))

# Create a count plot to visualize booking cancellations per month
ax = sns.countplot(x=df_copy['arrival_date_month'], hue=df_copy['is_canceled'],palette="mako",data=df_copy)

# Set title and axis labels
ax.set_title('Booking Distribution by Month')
ax.set_xlabel('Month')
ax.set_ylabel('Number of bookings')

# Set the x-axis ticks and labels
ax.set_xticks(range(12))

ax.set_xticklabels(['Jan', 'Feb', 'March', 'April', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec'])
# Show the plot
plt.show()
````

##### 1. Why did you pick the specific chart?



The count plot was chosen for its effectiveness in showcasing the distribution of booking cancellations across different months. Count plots excel at visually representing categorical data frequency, making them well-suited for highlighting bookings and cancellations per month.

Leveraging the "hue" parameter to distinguish between canceled and non-canceled bookings, this plot offers a visually appealing comparison in a single visualization, enabling easy recognition of patterns and trends.

##### 2. What is/are the insight(s) found from the chart?

The bar chart illustrates the following insights:


    
  * **Seasonal Booking Trends**: August and July exhibit peak bookings, indicating a busier travel or vacation season, with noticeable declines in November, February, and January.

  * **Cancellation Patterns**: Higher cancellations align with the peak booking months of August and July, while lower cancellation rates correspond to months with fewer total bookings, such as January and November.

  * **Correlation Between Bookings and Cancellations**: Months with higher bookings also exhibit a higher number of cancellations, suggesting a proportional relationship.

  * **Holiday Impact**: The peak booking months of August and July likely reflect the impact of vacations, potentially leading to higher cancellations due to changes in travel plans.
